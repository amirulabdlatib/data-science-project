# Student Perfomance Analysis (Math subject)

## 1.Data Preprocessing

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

Read the csv data and display from top

```python
df = pd.read_csv('student-mat.csv',sep=';')
df.head()
```

```python
#Check missing data
df.apply(lambda x: sum(x.isnull()),axis=0) 
```

```python
df.info()
```

```python
var_mod = ['school','sex','internet','address']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
df.dtypes
```

```python
df.head()
```

```python
df.describe()
```

## 2. Exploratory Data Analysis

What is the distribution of final grades (G3) in the Math course?

Describing the grade distribution.

```python
math_grades = df['G3']

plt.hist(math_grades,edgecolor='k',color='green')
plt.title('Math Grades distribution')
plt.xlabel('Marks')
plt.ylabel('Number of students')
plt.show()
```

```python
mean_grade = math_grades.mean()
median_grade = math_grades.median()
std_grade = math_grades.std()


print(f'The average marks for math test is: {mean_grade:.2f}')
print(f'The median marks for math test is: {median_grade:.2f}')
print(f'The standard deviation marks for math test is: {std_grade:.2f}')
```

```python
school_sex = df.groupby(['sex','school']).count()['G3']
school_sex 
```

```python
unstacked_school_sex = school_sex.unstack(level=0)
unstacked_school_sex 
```

```python
unstacked_school_sex.plot(kind='bar',stacked='True')
plt.show()
```

How does student gender (sex) correlate with their final grades (G3) in the Math course?

```python
corr_matrix = df.corr()
corr_matrix['sex']
```

The correlation between sex and their final grades 'G3' is 0.103456 which indicates they are having a weak positive coreration

Examining a simple correlation between gender and grades.
Is there an age-related pattern in math performance (G3) among students?

```python
corr_mat = df.corr().round(2)
corr_mat['age']
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Round the correlation coefficients to two decimal points
corr_matrix = corr_matrix.round(2)

# Create a heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Age has weak negative correlation with their final grades as it only have -0.16
